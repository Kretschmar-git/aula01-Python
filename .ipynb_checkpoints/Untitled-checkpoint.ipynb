```python
contMulheres = 0
somaAlturaMulheres = 0
for i in range(6):
    altura = float(input(f"{i+1} - Digite uma altura: "))
    sexo = input(f"{i+1} - Digite o sexo (M/F): ").upper()
    if(i == 0):
        menorAltura = altura
        maiorAltura = altura

    if(altura < menorAltura):
        menorAltura = altura

    if(altura > maiorAltura):
        maiorAltura = altura

    if(sexo == 'F'):
        contMulheres += 1
        somaAlturaMulheres += altura

media = somaAlturaMulheres / contMulheres

print(f"A menor altura é: {menorAltura}")
print(f"A maior altura é: {maiorAltura}")
print(f"A quantidade de mulheres é: {contMulheres}")
print(f"A média de altura das mulheres é: {media}")
```

```python
alts = []
altsF = []
for i in range(3):
    while True:
        try:
            altura = float(input(f"{i+1} - Digite uma altura: "))
            break
        except ValueError:
            print("Difite uma altura valida")

    while True:
            sexo = input(f"{i+1} - Digite o sexo (M/F): ").upper()
            if sexo in ['M', 'F']:
                break
            else:
                print("Digite um sexo valido!")

    alts.append(altura)

    if(sexo == 'F'):
        altsF.append(altura)



print(f"A menor altura é: {min(alts)}")
print(f"A maior altura é: {max(alts)}")
print(f"A quantidade de mulheres é: {len(altsF)}")
print(f"A média de altura das mulheres é: {sum(altsF)/len(altsF)}")
```
